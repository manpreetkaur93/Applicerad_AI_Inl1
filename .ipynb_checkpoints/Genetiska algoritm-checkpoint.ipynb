### Data förberedelse

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Läsa och bearbeta data
def preprocess_data(file_path):
    data = pd.read_csv(file_path)
    
    #"Straff" (negativt värde om deadline är försenad)
    data['Straff'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)
    
    # Score för prioritering av paket
    data['Score'] = (data['Förtjänst'] * 2) - (data['Straff'] / 5) + (1 / (data['Vikt'] + 1))
    
    # Sortera efter högst Score
    data = data.sort_values(by='Score', ascending=False).reset_index(drop=True)
    return data

def visualize_initial_data(data):
    plt.figure(figsize=(10, 6))
    plt.hist(data['Vikt'], bins=20, alpha=0.6, label='Vikt')      
    plt.hist(data['Förtjänst'], bins=20, alpha=0.6, label='Förtjänst')  
    plt.title('Fördelning av initial data')
    plt.xlabel('Värde')
    plt.ylabel('Frekvens')
    plt.legend()
    plt.grid(True)
    plt.show()

file_path = "lagerstatus.csv"  
data = preprocess_data(file_path)
visualize_initial_data(data)

print("Sammanfattning av data:")
print(data.describe())


```

## Datahantering och Visualisering
#### Jag läser in filen `lagerstatus.csv`, beräknar straffavgifter för försenade paket och skapar en enkel score för att prioritera vilka paket som ska skickas först. Därefter sorteras datan och jag plottar ett histogram för att få en överblick av vikt och förtjänst.


```python
# Justera poäng baserat på straff och deadline
data['Justerad_poäng'] = (
    data['Score'] 
    + (data['Straff'] * -0.5) 
    + (10 / (data['Deadline'].abs() + 1))
)

# Sortera på nytt
data = data.sort_values(by=['Straff', 'Justerad_poäng'], ascending=[True, False])

# Visualisera deadline-fördelning
plt.figure(figsize=(10, 6))
plt.hist(data['Deadline'], bins=20, color='purple', alpha=0.7, edgecolor='black')
plt.title('Deadline-fördelning')
plt.xlabel('Deadline (dagar)')
plt.ylabel('Frekvens')
plt.grid()
plt.show()

# Visualisera straffavgift mot förtjänst
plt.figure(figsize=(10, 6))
plt.scatter(data['Straff'], data['Förtjänst'], alpha=0.5, c=data['Deadline'], cmap='viridis')
plt.colorbar(label='Deadline')
plt.title('Straffavgift vs Förtjänst')
plt.xlabel('Straffavgift')
plt.ylabel('Förtjänst')
plt.grid()
plt.show()
```

## Poängjustering och Deadline-Visualisering
#### Här justerar jag poäng enligt hur sent ett paket är och sorterar därefter paketen så att de viktigaste hanteras först. Jag plottar också hur deadlines och straffavgifter fördelar sig för att få en tydlig översikt.


```python
import os  

# Konstanter
VEHICLE_CAPACITY = 800
NUM_VEHICLES = 10
GENERATIONS = 10
POPULATION_SIZE = 5
MUTATION_COUNT = 5


# Läser in data i chunkar och beräknar en straffavgift för paket som har deadline redan passerat. 
def load_large_data(file_path, chunksize=10000):
    data_chunks = pd.read_csv(file_path, chunksize=chunksize)
    data = pd.concat(data_chunks, ignore_index=True)
    data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)
    return data

# Score som väger in förtjänst, straffavgift och vikt och försenade paket får extra prioritet
def allocate_packages_with_deadline_policy(data):
    data['Score'] = (
        (data['Förtjänst'] * 2)
        - (data['Penalty'] / 3)
        + (1 / (data['Vikt'] + 1))
        + ((data['Deadline'] < 0) * 5)
    )
    data = data.sort_values(by='Score', ascending=False)

    vehicles = [[] for _ in range(NUM_VEHICLES)]
    vehicle_weights = [0] * NUM_VEHICLES
    package_count_map = {pkg_id: 0 for pkg_id in data['Paket_id']}

    # Försöker placera paket i första fordon som har plats
    for _, pkg in data.iterrows():
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg['Vikt'] <= VEHICLE_CAPACITY:
                vehicles[i].append(pkg['Paket_id'])
                vehicle_weights[i] += pkg['Vikt']
                package_count_map[pkg['Paket_id']] += 1
                break
    data['Loaded'] = data['Paket_id'].apply(lambda x: 1 if package_count_map[x] > 0 else 0)
    return vehicles, data

def fitness_score(data, vehicles):
    """Beräknar fitness = (total förtjänst för lastade paket) - (straffavgift för ej lastade). Returnerar även antalet försenade paket som levererats, 
    samt försenade paket kvar i lager."""
    assigned_packages = set(pkg for vehicle in vehicles for pkg in vehicle)
    unallocated_packages = data[~data['Paket_id'].isin(assigned_packages)]
    
    delayed_delivered = len(data[(data['Paket_id'].isin(assigned_packages)) & (data['Deadline'] < 0)])
    delayed_stock = len(unallocated_packages[unallocated_packages['Deadline'] < 0])
    
    total_profit = data[data['Paket_id'].isin(assigned_packages)]['Förtjänst'].sum()
    total_penalty = unallocated_packages['Penalty'].sum()

    return total_profit - total_penalty, total_penalty, delayed_delivered, delayed_stock

# Mutation
def optimize_mutation(vehicles, data):
    """Flyttar slumpmässigt ett paket från ett fordon till ett annat om det finns plats. """
    mutated_vehicles = [list(vehicle) for vehicle in vehicles]
    vehicle_weights = [
        sum(data[data['Paket_id'].isin(vehicle)]['Vikt']) for vehicle in mutated_vehicles
    ]

    for _ in range(MUTATION_COUNT):
        vehicle_from = random.randint(0, NUM_VEHICLES - 1)
        vehicle_to = random.randint(0, NUM_VEHICLES - 1)

        if mutated_vehicles[vehicle_from]:
            pkg_id = random.choice(mutated_vehicles[vehicle_from])
            pkg_weight = data.loc[data['Paket_id'] == pkg_id, 'Vikt'].values[0]

            if vehicle_weights[vehicle_to] + pkg_weight <= VEHICLE_CAPACITY:
                mutated_vehicles[vehicle_from].remove(pkg_id)
                mutated_vehicles[vehicle_to].append(pkg_id)
                vehicle_weights[vehicle_from] -= pkg_weight
                vehicle_weights[vehicle_to] += pkg_weight

    return mutated_vehicles

# Genetisk Algoritm med stoppkriterium
def genetic_algorithm_with_stopping(data, max_generations=GENERATIONS, tolerance=0.01, patience=5):
    """Skapar en population av lösningar, utvärderar fitness och muterar. """
    # Generera startpopulation
    population = [allocate_packages_with_deadline_policy(data)[0] for _ in range(POPULATION_SIZE)]
    
    best_solution = None
    best_fitness = float('-inf')
    
    best_total_penalty = 0
    best_delayed_delivered = 0
    best_delayed_stock = 0

    fitness_progress = []
    stagnant_generations = 0

    for generation in range(max_generations):
        fitness_scores = []
        for vehicles in population:
            fit, total_penalty, delayed_delivered, delayed_stock = fitness_score(data, vehicles)
            fitness_scores.append((fit, vehicles, total_penalty, delayed_delivered, delayed_stock))

        # Sortera efter högst fitness
        fitness_scores.sort(reverse=True, key=lambda x: x[0])

        # Om bästa fitness förbättrats då uppdatera
        if fitness_scores[0][0] > best_fitness:
            best_fitness = fitness_scores[0][0]
            best_solution = fitness_scores[0][1]
            best_total_penalty = fitness_scores[0][2]
            best_delayed_delivered = fitness_scores[0][3]
            best_delayed_stock = fitness_scores[0][4]
            stagnant_generations = 0
        else:
            stagnant_generations += 1

        fitness_progress.append(best_fitness)

        # Avbryt om ingen förbättring skett 
        if stagnant_generations >= patience:
            print(f"Algoritmen stannade vid generation {generation + 1} p.g.a. låg förbättring.")
            break

        # Välj topp hälften och mutera för nästa generation
        next_generation = [fitness_scores[i][1] for i in range(POPULATION_SIZE // 2)]
        for i in range(0, len(next_generation), 2):
            if i + 1 < len(next_generation):
                parent1, parent2 = next_generation[i], next_generation[i + 1]
                offspring1 = optimize_mutation(parent1, data)
                offspring2 = optimize_mutation(parent2, data)
                next_generation.extend([offspring1, offspring2])

        # Begränsa populationen
        population = next_generation[:POPULATION_SIZE]

    return best_solution, best_fitness, fitness_progress, best_total_penalty, best_delayed_delivered, best_delayed_stock


# Spara resultat till CSV
def save_original_assignment(data, output_file="optimized_vehicle_allocation.csv"):
    data_copy = data.copy()
    if not os.path.exists(output_file):
        data_copy.to_csv(output_file, index=False)
        print(f"Original data saved to '{output_file}'")
    else:
        print(f"'{output_file}' already exists. Overwriting...")
        data_copy.to_csv(output_file, index=False)

def save_assigned_vehicles_summary(data, vehicles, output_file="assigned_vehicles_summary.csv"):
    summary = []
    for vehicle_id, vehicle_list in enumerate(vehicles, start=1):
        vehicle_packages = data[data['Paket_id'].isin(vehicle_list)]
        total_weight = vehicle_packages['Vikt'].sum()
        num_packages = len(vehicle_list)
        total_profit = vehicle_packages['Förtjänst'].sum()

        summary.append({
            "Vehicle_ID": vehicle_id,
            "Total_Weight": round(total_weight, 2),
            "Number_of_Packages": num_packages,
            "Total_Profit": round(total_profit, 2)
        })

    summary_df = pd.DataFrame(summary)
    if not os.path.exists(output_file):
        summary_df.to_csv(output_file, index=False)
        print(f"Assigned vehicles summary saved to '{output_file}'")
    else:
        print(f"'{output_file}' already exists. Overwriting...")
        summary_df.to_csv(output_file, index=False)

# Visualiseringsfunktioner
def visualize_fitness_progress(fitness_progress):
    plt.figure(figsize=(8, 5))
    plt.plot(fitness_progress, marker='o', linestyle='-')
    plt.title("Fitnessutveckling över generationer")
    plt.xlabel("Generation")
    plt.ylabel("Fitness")
    plt.grid(True)
    plt.show()

def visualize_vehicle_allocation(best_solution, data, best_fitness, total_penalty, delayed_delivered, delayed_stock):
    vehicle_loads = [len(vehicle) for vehicle in best_solution]
    
    plt.figure(figsize=(8, 5))
    plt.bar(range(1, len(vehicle_loads) + 1), vehicle_loads, color='skyblue', edgecolor='black')
    plt.title("Antal paket per fordon")
    plt.xlabel("Fordon")
    plt.ylabel("Antal paket")
    plt.grid(True)
    plt.show()
    
    print("Bästa fitnessvärde:", best_fitness)
    print("Total straffavgift för oallokerade paket:", total_penalty)
    print("Antal försenade men levererade paket:", delayed_delivered)
    print("Antal försenade kvar i lager:", delayed_stock)

# Kör algoritmen och spara resultat
if __name__ == "__main__":
    data = load_large_data("lagerstatus.csv")
    vehicles, updated_data = allocate_packages_with_deadline_policy(data)

    #genetisk algoritm
    best_solution, best_fitness, fitness_progress, total_penalty, delayed_delivered, delayed_stock = \
        genetic_algorithm_with_stopping(data)

    visualize_fitness_progress(fitness_progress)
    visualize_vehicle_allocation(best_solution, data, best_fitness, total_penalty, delayed_delivered, delayed_stock)
    
    save_original_assignment(updated_data, output_file="optimized_vehicle_allocation.csv")
    save_assigned_vehicles_summary(updated_data, vehicles, output_file="assigned_vehicles_summary.csv")
```

## Genetisk Algoritm 
Jag läser först in filen `lagerstatus.csv` och beräknar en straffavgift (Penalty) för de paket som är försenade. 

Sedan använder jag en genetisk algoritm för att iterera fram en bättre fördelning. Den förbättrar fördelningen successivt genom mutationer och stannar om den inte längre får bättre resultat.

- **Antal paket per fordon** visar hur fordons belastning.

### Paket kvar i lager

```python
def evaluate_unallocated_packages(data, vehicles):
    assigned_packages = set(pkg for vehicle in vehicles for pkg in vehicle)
    data['Loaded'] = data['Paket_id'].apply(lambda x: 1 if x in assigned_packages else 0)

    unallocated_packages = data[data['Loaded'] == 0]
    leftover_count = len(unallocated_packages)
    leftover_profit = unallocated_packages['Förtjänst'].sum()

    return leftover_count, leftover_profit, unallocated_packages

def visualize_unallocated_packages(unallocated_packages):
    plt.figure(figsize=(12, 6))
    plt.hist(unallocated_packages['Vikt'], bins=20, color='purple', alpha=0.8, label="Vikt")
    plt.title("Vikt för paket kvar i lager")
    plt.xlabel("Vikt (kg)")
    plt.ylabel("Frekvens")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    plt.figure(figsize=(12, 6))
    plt.hist(unallocated_packages['Förtjänst'], bins=20, color='green', alpha=0.8, label="Förtjänst")
    plt.title("Förtjänst för paket kvar i lager")
    plt.xlabel("Förtjänst")
    plt.ylabel("Frekvens")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

def calculate_statistics(df, column, title="Statistik"):
    values = df[column]
    print(title)
    print(f"  Antal: {len(values)}")
    print(f"  Medelvärde: {values.mean():.2f}")
    print(f"  Varians: {values.var():.2f}")
    print(f"  Standardavvikelse: {values.std():.2f}")
    print(f"  Min: {values.min()}")
    print(f"  Max: {values.max()}")
    print("--------------------------------------------------")


leftover_count, leftover_profit, unallocated_packages = evaluate_unallocated_packages(data, vehicles)

print(f"Antal paket kvar i lager: {leftover_count}")
print(f"Total förtjänst kvar i lager (exkl. straff): {leftover_profit}")

# Visualisera
visualize_unallocated_packages(unallocated_packages)

# Statistisk analys
calculate_statistics(unallocated_packages, 'Vikt', "Statistik för vikt kvar i lager")
calculate_statistics(unallocated_packages, 'Förtjänst', "Statistik för förtjänst kvar i lager")
```
