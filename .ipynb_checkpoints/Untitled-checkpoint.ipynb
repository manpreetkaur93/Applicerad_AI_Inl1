```python
# Import necessary libraries
import pandas as pd

# Define file paths (replace with your file paths if different)
file_paths = [
    'lagerstatus.csv',
    'lagerstatus2.csv',
    'lagerstatus3.csv',
    'lagerstatus4.csv'
]

# Load data from all files into a list of DataFrames
data_frames = [pd.read_csv(file) for file in file_paths]

# Concatenate all DataFrames into one
all_data = pd.concat(data_frames, ignore_index=True)

# Display basic information about the dataset
print("Dataset Information:")
print(all_data.info())

# Display the first few rows of the dataset
print("\nFirst 5 rows of the dataset:")
print(all_data.head())

# Check for missing values
print("\nMissing Values Check:")
print(all_data.isnull().sum())

# Validate data types
print("\nData Types:")
print(all_data.dtypes)
```

```python
import numpy as np

# Filter out invalid rows if necessary (e.g., negative or zero weight)
filtered_data = all_data[all_data['Vikt'] > 0]

# Add a penalty column for overdue packages
filtered_data['Penalty'] = filtered_data['Deadline'].apply(
    lambda x: 0 if x >= 0 else x**2
)

# Calculate net profit (profit - penalty)
filtered_data['Net_Profit'] = filtered_data['Förtjänst'] - filtered_data['Penalty']

# Display basic statistics
print("\nBasic Statistics:")
print(f"Total Packages: {len(filtered_data)}")
print(f"Total Weight: {filtered_data['Vikt'].sum():.2f} kg")
print(f"Total Profit: {filtered_data['Förtjänst'].sum()}")
print(f"Total Penalty: {filtered_data['Penalty'].sum()}")

# Display summary of the prepared data
print("\nPrepared Data (First 5 Rows):")
print(filtered_data.head())
```

""" Now that the data is clean and prepared, we can move on to designing the optimization process using a genetic algorithm (GA). This algorithm will help us assign packages to vans in a way that maximizes profit while staying within weight constraints."""

### Simplified Genetic Algorithm
###### A genetic algorithm will help us decide which packages go into which van to maximize profit. Here's how we'll make it easier:

1. Key Concepts:

###### Each solution represents a way to assign packages to vans.
###### We test how "good" each solution is using a fitness function:
###### Good Solution = High profit, vans stay within weight limits.
###### We improve solutions over time using random changes (mutation) and mixing solutions (crossover).
2. Simple Plan:

###### Start with random assignments of packages to vans.
###### Test how well each assignment works (fitness).
###### Keep improving the assignments over several iterations (generations).

```python
import random

# Constants
NUM_VANS = 10  # Number of vans
VAN_CAPACITY = 800  # Max weight per van
POPULATION_SIZE = 20  # Number of solutions to test at once

# Step 1: Initialize Valid Population
def initialize_valid_population_optimized(data, num_vans, van_capacity, population_size):
    """
    Creates an initial population with valid solutions using a greedy strategy.
    Ensures no van exceeds its weight limit.
    """
    population = []
    
    # Sort packages by weight (heaviest first)
    sorted_packages = data.sort_values(by='Vikt', ascending=False).index.tolist()
    
    for _ in range(population_size):
        solution = [[] for _ in range(num_vans)]  # Empty vans
        van_weights = [0] * num_vans  # Track current weight of each van

        for package_idx in sorted_packages:
            package_weight = data.loc[package_idx, 'Vikt']
            
            # Assign the package to the first van that can take it
            for van_idx in range(num_vans):
                if van_weights[van_idx] + package_weight <= van_capacity:
                    solution[van_idx].append(package_idx)
                    van_weights[van_idx] += package_weight
                    break  # Move to the next package
        
        population.append(solution)
    return population

# Step 2: Fitness Function
def calculate_fitness(solution, data):
    """
    Calculates the total profit for a solution. Returns -1 if any van exceeds its weight limit.
    """
    total_profit = 0
    for van in solution:
        van_weight = sum(data.loc[van, 'Vikt'])
        van_profit = sum(data.loc[van, 'Net_Profit'])
        
        # Check if van is overweight
        if van_weight > VAN_CAPACITY:
            return -1  # Invalid solution if any van exceeds weight
        total_profit += van_profit
    return total_profit

# Step 3: Evaluate Population
def evaluate_population(population, data):
    """
    Evaluates the fitness of each solution in the population.
    """
    return [calculate_fitness(solution, data) for solution in population]

# Run Initialization and Evaluation
population = initialize_valid_population_optimized(filtered_data, NUM_VANS, VAN_CAPACITY, POPULATION_SIZE)
fitness_scores = evaluate_population(population, filtered_data)

# Display Results
print("Initial Population Fitness:")
print(f"Best Fitness (Profit): {max(fitness_scores)}")
print(f"Average Fitness: {sum(fitness_scores) / len(fitness_scores):.2f}")

```

### The initial population is now valid, and we have a solid foundation to improve further. With a Best Fitness of 2465 and a valid initial population, we can move to the next step: Mutation to refine the solutions.

###### Next Step: Implement and Apply Mutation
' Mutation Purpose:

##### Make small, random changes to the solutions to explore new possibilities and potentially improve the overall fitness.


### simplified mutation code snippet to fix and improve invalid solutions by randomly tweaking them.

```python
def mutate_solution(solution, data, num_vans):
    """
    Applies mutation by randomly moving packages between vans to improve the solution.
    Ensures no van exceeds the weight limit after mutation.
    """
    # Randomly pick a van and a package from that van
    van_from = random.randint(0, num_vans - 1)
    if len(solution[van_from]) == 0:  # Skip if the van is empty
        return solution  # No mutation applied
    
    package_idx = random.choice(solution[van_from])  # Select a random package
    
    # Randomly pick another van to move the package to
    van_to = random.randint(0, num_vans - 1)
    while van_to == van_from:  # Ensure it's not the same van
        van_to = random.randint(0, num_vans - 1)

    # Temporarily move the package
    solution[van_from].remove(package_idx)
    solution[van_to].append(package_idx)

    # Check if mutation is valid (i.e., weight constraints are met)
    van_from_weight = sum(data.loc[solution[van_from], 'Vikt'])
    van_to_weight = sum(data.loc[solution[van_to], 'Vikt'])

    if van_from_weight > VAN_CAPACITY or van_to_weight > VAN_CAPACITY:
        # Revert mutation if it violates weight constraints
        solution[van_to].remove(package_idx)
        solution[van_from].append(package_idx)

    return solution  # Return the mutated solution

# Apply mutation to the population
def mutate_population(population, data, num_vans):
    """
    Mutates the entire population to explore new solutions.
    """
    mutated_population = [mutate_solution(solution, data, num_vans) for solution in population]
    return mutated_population

# Mutate the population
mutated_population = mutate_population(population, filtered_data, NUM_VANS)

# Evaluate the mutated population
mutated_fitness_scores = evaluate_population(mutated_population, filtered_data)

# Show Results
print("\nMutated Population Fitness:")
print(f"Best Fitness (Profit): {max(mutated_fitness_scores)}")
print(f"Average Fitness: {sum(mutated_fitness_scores) / len(mutated_fitness_scores):.2f}")
```

<h3> The mutation step did not lead to an improvement in fitness, which suggests the current solutions are already near-optimal or that the mutation is too conservative to explore new possibilities. This is normal in genetic algorithms and is why we combine mutation with crossover.

<h3> Next Step: Implement and Apply Crossover
<h3>  Crossover Purpose:

<h3> Combine two good solutions (parents) to create a new solution (child).
Helps explore new combinations of package assignments, potentially improving fitness.

```python
def crossover(parent1, parent2, num_vans):
    """
    Combines two solutions (parents) to create a new solution (child).
    Each van in the child is randomly inherited from either parent.
    """
    child = [[] for _ in range(num_vans)]
    
    for van_idx in range(num_vans):
        # Randomly pick vans from either parent
        if random.random() < 0.5:
            child[van_idx] = parent1[van_idx].copy()
        else:
            child[van_idx] = parent2[van_idx].copy()
    
    return child

def crossover_population(population, num_vans):
    """
    Applies crossover to the entire population to generate new solutions.
    Pairs solutions randomly for crossover.
    """
    new_population = []
    for _ in range(len(population) // 2):  # Pair up solutions
        parent1 = random.choice(population)
        parent2 = random.choice(population)
        
        # Generate two children from the parents
        child1 = crossover(parent1, parent2, num_vans)
        child2 = crossover(parent2, parent1, num_vans)
        
        new_population.extend([child1, child2])
    
    return new_population

# Apply crossover to the population
crossover_population_result = crossover_population(mutated_population, NUM_VANS)

# Evaluate the population after crossover
crossover_fitness_scores = evaluate_population(crossover_population_result, filtered_data)

# Show Results
print("\nCrossover Population Fitness:")
print(f"Best Fitness (Profit): {max(crossover_fitness_scores)}")
print(f"Average Fitness: {sum(crossover_fitness_scores) / len(crossover_fitness_scores):.2f}")
```

When the fitness scores remain unchanged, it typically indicates that the algorithm has reached a local optimum or that the exploration mechanisms (mutation and crossover) aren't diverse enough to generate better solutions.

Reasons for No Improvement:
Already Optimal Solution:
The current best solution might already be the best achievable within the constraints.
Limited Exploration:
Mutation and crossover might not be introducing enough diversity to explore new solutions.
Population Stagnation:
The population might lack diversity, so mutation and crossover aren't creating significantly different solutions.


```python
import random

# Updated mutation functions

# 1. Mutate by moving multiple packages
def mutate_solution_with_multiple_changes(solution, data, num_vans, num_changes=3):
    """
    Mutates a solution by randomly moving multiple packages between vans.
    """
    for _ in range(num_changes):  # Perform multiple mutations
        van_from = random.randint(0, num_vans - 1)
        if len(solution[van_from]) == 0:  # Skip empty vans
            continue
        
        package_idx = random.choice(solution[van_from])  # Pick a package
        van_to = random.randint(0, num_vans - 1)
        while van_to == van_from:  # Ensure it's a different van
            van_to = random.randint(0, num_vans - 1)

        # Move the package
        solution[van_from].remove(package_idx)
        solution[van_to].append(package_idx)

        # Check for weight constraints
        van_from_weight = sum(data.loc[solution[van_from], 'Vikt'])
        van_to_weight = sum(data.loc[solution[van_to], 'Vikt'])
        if van_from_weight > VAN_CAPACITY or van_to_weight > VAN_CAPACITY:
            # Revert if invalid
            solution[van_to].remove(package_idx)
            solution[van_from].append(package_idx)
    
    return solution

# 2. Mutate by swapping packages between vans
def mutate_solution_with_swaps(solution, data, num_vans, num_swaps=2):
    """
    Swaps packages between two random vans to increase diversity.
    """
    for _ in range(num_swaps):  # Perform multiple swaps
        van1 = random.randint(0, num_vans - 1)
        van2 = random.randint(0, num_vans - 1)
        if van1 == van2 or len(solution[van1]) == 0 or len(solution[van2]) == 0:
            continue  # Skip invalid swaps

        # Swap one random package between vans
        package1 = random.choice(solution[van1])
        package2 = random.choice(solution[van2])

        solution[van1].remove(package1)
        solution[van2].remove(package2)
        solution[van1].append(package2)
        solution[van2].append(package1)

        # Check weight constraints and revert if invalid
        van1_weight = sum(data.loc[solution[van1], 'Vikt'])
        van2_weight = sum(data.loc[solution[van2], 'Vikt'])
        if van1_weight > VAN_CAPACITY or van2_weight > VAN_CAPACITY:
            # Revert swaps
            solution[van1].remove(package2)
            solution[van2].remove(package1)
            solution[van1].append(package1)
            solution[van2].append(package2)
    
    return solution

# 3. Apply a mutation strategy to the population
def mutate_population_with_diversity(population, data, num_vans, mutation_rate=0.5):
    """
    Mutates a fraction of the population using different mutation strategies.
    """
    mutated_population = []
    for solution in population:
        if random.random() < mutation_rate:  # Apply mutation with a probability
            if random.random() < 0.5:
                # Apply multiple changes
                mutated_population.append(mutate_solution_with_multiple_changes(solution, data, num_vans))
            else:
                # Apply swaps
                mutated_population.append(mutate_solution_with_swaps(solution, data, num_vans))
        else:
            # Keep solution unchanged
            mutated_population.append(solution)
    
    # Add one or more random solutions to increase diversity
    random_solution = initialize_valid_population_optimized(data, num_vans, VAN_CAPACITY, 1)[0]
    mutated_population.append(random_solution)
    
    return mutated_population

# Apply mutation to the population
mutated_population = mutate_population_with_diversity(population, filtered_data, NUM_VANS, mutation_rate=0.6)

# Evaluate the mutated population
mutated_fitness_scores = evaluate_population(mutated_population, filtered_data)

# Show Results
print("\nMutated Population Fitness with Diversity:")
print(f"Best Fitness (Profit): {max(mutated_fitness_scores)}")
print(f"Average Fitness: {sum(mutated_fitness_scores) / len(mutated_fitness_scores):.2f}")
```

```python
import matplotlib.pyplot as plt

# Initialize lists to track progress
best_fitness_history = []
average_fitness_history = []

# Parameters
NUM_GENERATIONS = 20

# Run the optimization over multiple generations
current_population = population  # Start with the initial population
for generation in range(NUM_GENERATIONS):
    # Apply mutation
    mutated_population = mutate_population_with_diversity(current_population, filtered_data, NUM_VANS, mutation_rate=0.6)
    
    # Apply crossover
    crossover_population_result = crossover_population(mutated_population, NUM_VANS)
    
    # Evaluate fitness
    fitness_scores = evaluate_population(crossover_population_result, filtered_data)
    
    # Track best and average fitness
    best_fitness_history.append(max(fitness_scores))
    average_fitness_history.append(sum(fitness_scores) / len(fitness_scores))
    
    # Update the current population for the next generation
    current_population = crossover_population_result

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_history, label='Best Fitness', marker='o')
plt.plot(average_fitness_history, label='Average Fitness', marker='x')
plt.title('Optimization Progress Over Generations')  # Complete the string
plt.xlabel('Generation')
plt.ylabel('Fitness (Profit)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
def initialize_valid_population_prioritize_deadlines(data, num_vans, van_capacity, population_size):
    """
    Creates an initial population with valid solutions using a greedy strategy.
    Packages with earlier deadlines are prioritized to reduce penalties.
    """
    population = []
    
    # Sort packages by deadline (earliest first), then by weight (heaviest first)
    sorted_packages = data.sort_values(by=['Deadline', 'Vikt'], ascending=[True, False]).index.tolist()
    
    for _ in range(population_size):
        solution = [[] for _ in range(num_vans)]  # Empty vans
        van_weights = [0] * num_vans  # Track current weight of each van

        for package_idx in sorted_packages:
            package_weight = data.loc[package_idx, 'Vikt']
            
            # Assign the package to the first van that can take it
            for van_idx in range(num_vans):
                if van_weights[van_idx] + package_weight <= van_capacity:
                    solution[van_idx].append(package_idx)
                    van_weights[van_idx] += package_weight
                    break  # Move to the next package
        
        population.append(solution)
    return population

# Re-initialize the population with the updated strategy
population = initialize_valid_population_prioritize_deadlines(filtered_data, NUM_VANS, VAN_CAPACITY, POPULATION_SIZE)

# Evaluate the new population
fitness_scores = evaluate_population(population, filtered_data)

# Display Results
print("\nUpdated Population Fitness (Prioritize Deadlines):")
print(f"Best Fitness (Profit): {max(fitness_scores)}")
print(f"Average Fitness: {sum(fitness_scores) / len(fitness_scores):.2f}")

# Visualize Delivery Results (same as before)
best_solution_index = np.argmax(fitness_scores)
best_solution = population[best_solution_index]

van_stats, total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(best_solution, filtered_data)

# Print Updated Results
print("\nUpdated Delivery Results:")
print(f"Total Profit for Deliveries: {total_profit}")
print(f"Total Penalty for Overdue Packages: {total_penalty}")
print(f"Number of Packages Left in Warehouse: {leftover_count}")
print(f"Total Weight of Leftover Packages: {leftover_weight}")
print(f"Total Profit of Leftover Packages: {leftover_profit}")
```

```python
def initialize_valid_population_hybrid_strategy(data, num_vans, van_capacity, population_size):
    """
    Creates an initial population using a hybrid strategy that balances profit and deadlines.
    """
    # Calculate a priority metric: Higher profit + earlier deadline
    data['Profit_Weight'] = data['Förtjänst'] / (1 + abs(data['Deadline']))  # Smaller deadlines are better
    sorted_packages = data.sort_values(by=['Profit_Weight', 'Vikt'], ascending=[False, False]).index.tolist()

    population = []
    for _ in range(population_size):
        solution = [[] for _ in range(num_vans)]  # Empty vans
        van_weights = [0] * num_vans  # Track current weight of each van

        for package_idx in sorted_packages:
            package_weight = data.loc[package_idx, 'Vikt']

            # Assign the package to the first van that can take it
            for van_idx in range(num_vans):
                if van_weights[van_idx] + package_weight <= van_capacity:
                    solution[van_idx].append(package_idx)
                    van_weights[van_idx] += package_weight
                    break  # Move to the next package

        population.append(solution)
    return population

# Re-initialize population using hybrid strategy
population = initialize_valid_population_hybrid_strategy(filtered_data, NUM_VANS, VAN_CAPACITY, POPULATION_SIZE)

# Evaluate the new population
fitness_scores = evaluate_population(population, filtered_data)

# Display Results
print("\nUpdated Population Fitness (Hybrid Strategy):")
print(f"Best Fitness (Profit): {max(fitness_scores)}")
print(f"Average Fitness: {sum(fitness_scores) / len(fitness_scores):.2f}")

# Visualize Delivery Results (same as before)
best_solution_index = np.argmax(fitness_scores)
best_solution = population[best_solution_index]

van_stats, total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(best_solution, filtered_data)

# Print Updated Results
print("\nUpdated Delivery Results (Hybrid Strategy):")
print(f"Total Profit for Deliveries: {total_profit}")
print(f"Total Penalty for Overdue Packages: {total_penalty}")
print(f"Number of Packages Left in Warehouse: {leftover_count}")
print(f"Total Weight of Leftover Packages: {leftover_weight}")
print(f"Total Profit of Leftover Packages: {leftover_profit}")
```

```python
# Updated Fitness Function
def calculate_fitness_hybrid(solution, data):
    """
    Calculates fitness by balancing profit and penalty.
    Returns net profit as the fitness score.
    """
    total_profit = 0
    total_penalty = 0

    for van in solution:
        van_profit = sum(data.loc[van, 'Förtjänst'])
        van_penalty = sum(data.loc[van, 'Penalty'])

        # Add van profit and penalties
        total_profit += van_profit
        total_penalty += van_penalty

    # Net profit = profit - penalty
    net_profit = total_profit - total_penalty
    return net_profit

# Apply new fitness function to evaluate the population
fitness_scores = [calculate_fitness_hybrid(solution, filtered_data) for solution in population]

# Display Results
print("\nUpdated Fitness Scores (Hybrid Fitness Function):")
print(f"Best Fitness (Net Profit): {max(fitness_scores)}")
print(f"Average Fitness: {sum(fitness_scores) / len(fitness_scores):.2f}")

# Display Delivery Results with New Fitness Function
best_solution_index = np.argmax(fitness_scores)
best_solution = population[best_solution_index]

van_stats, total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(best_solution, filtered_data)

print("\nUpdated Delivery Results (Refined Fitness Function):")
print(f"Total Profit for Deliveries: {total_profit}")
print(f"Total Penalty for Overdue Packages: {total_penalty}")
print(f"Number of Packages Left in Warehouse: {leftover_count}")
print(f"Total Weight of Leftover Packages: {leftover_weight}")
print(f"Total Profit of Leftover Packages: {leftover_profit}")
```

### How to better utilize van capacity?
##### Prioritize by Weight Efficiency
Assign packages with higher profit per kilogram to maximize profit per unit of capacity.
Sort packages by Profit_Per_Weight before assigning them to vans:
Greedy Strategy for Van Assignment
After sorting by Profit_Per_Weight, assign packages to vans in a way that tries to fully utilize the remaining space:
First, assign packages that perfectly fit the remaining capacity.
Then, fill remaining space with smaller packages.
Use a Knapsack Approach
Treat each van as a "knapsack" with a maximum capacity of 800 kg.
Solve the knapsack problem for each van to maximize profit while fully utilizing the available capacity.

```python
def knapsack_greedy(van_capacity, packages, data):
    """
    Solves a simplified knapsack problem for a single van.
    """
    selected = []  # List to store selected package indices
    total_weight = 0
    total_profit = 0

    for package_idx in packages:
        package_weight = data.loc[package_idx, 'Vikt']
        package_profit = data.loc[package_idx, 'Förtjänst']

        # Check if the package can fit in the remaining capacity
        if total_weight + package_weight <= van_capacity:
            selected.append(package_idx)  # Add package index to the list
            total_weight += package_weight
            total_profit += package_profit

    return selected, total_weight, total_profit

def assign_packages_with_knapsack(data, num_vans, van_capacity):
    """
    Assigns packages to vans using a greedy knapsack approach.
    """
    # Sort packages by Profit_Per_Weight
    data['Profit_Per_Weight'] = data['Förtjänst'] / data['Vikt']
    sorted_packages = data.sort_values(by='Profit_Per_Weight', ascending=False).index.tolist()

    solution = [[] for _ in range(num_vans)]  # Initialize vans as lists of lists
    remaining_packages = sorted_packages.copy()

    for van_idx in range(num_vans):
        selected, total_weight, total_profit = knapsack_greedy(van_capacity, remaining_packages, data)
        solution[van_idx] = list(selected)  # Ensure it's a list

        # Remove selected packages from the pool
        remaining_packages = [pkg for pkg in remaining_packages if pkg not in selected]

    return solution
```

```python
# Updated Fitness Function
def calculate_statistics(best_solution, data):
    """
    Calculates statistics for the best solution, including total profit, penalties,
    leftover packages, and more.
    """
    van_stats = []  # To store stats for each van
    total_profit = 0
    total_penalty = 0
    leftover_packages = data.index.tolist()  # Start with all packages in the warehouse

    for van in best_solution:
        if not isinstance(van, list):  # Ensure `van` is a list
            van = [van]

        van_weight = sum(data.loc[van, 'Vikt'])
        van_profit = sum(data.loc[van, 'Förtjänst'])
        van_penalty = sum(data.loc[van, 'Penalty'])

        total_profit += van_profit
        total_penalty += van_penalty
        van_stats.append({'weight': van_weight, 'profit': van_profit, 'penalty': van_penalty})

        # Remove assigned packages from the warehouse
        leftover_packages = [pkg for pkg in leftover_packages if pkg not in van]

    # Stats for leftover packages
    leftover_weight = sum(data.loc[leftover_packages, 'Vikt'])
    leftover_profit = sum(data.loc[leftover_packages, 'Förtjänst'])

    return van_stats, total_profit, total_penalty, len(leftover_packages), leftover_weight, leftover_profit
```

```python
# Assign packages using knapsack strategy
population = assign_packages_with_knapsack(filtered_data, NUM_VANS, VAN_CAPACITY)

# Evaluate and display results
fitness_scores = evaluate_population(population, filtered_data)
best_solution_index = np.argmax(fitness_scores)
best_solution = population[best_solution_index]

van_stats, total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(best_solution, filtered_data)

print("\nKnapsack Strategy Results:")
print(f"Total Profit for Deliveries: {total_profit}")
print(f"Total Penalty for Overdue Packages: {total_penalty}")
print(f"Number of Packages Left in Warehouse: {leftover_count}")
print(f"Total Weight of Leftover Packages: {leftover_weight}")
print(f"Total Profit of Leftover Packages: {leftover_profit}")
```

```python
import random

# Initialize Population with Knapsack Strategy
def initialize_population_knapsack(data, num_vans, van_capacity, population_size):
    population = []
    for _ in range(population_size):
        solution = assign_packages_with_knapsack(data, num_vans, van_capacity)
        population.append(solution)
    return population

# Evaluate Fitness (Net Profit)
def calculate_fitness(solution, data):
    total_profit = 0
    total_penalty = 0

    for van in solution:
        if not isinstance(van, list):
            van = [van]
        van_profit = sum(data.loc[van, 'Förtjänst'])
        van_penalty = sum(data.loc[van, 'Penalty'])

        total_profit += van_profit
        total_penalty += van_penalty

    return total_profit - total_penalty

# Selection Function
def select_parents(population, fitness_scores):
    sorted_population = [x for _, x in sorted(zip(fitness_scores, population), key=lambda pair: pair[0], reverse=True)]
    return sorted_population[:2]  # Select top 2 parents

# Crossover Function
def crossover(parent1, parent2, num_vans):
    child = [[] for _ in range(num_vans)]
    for i in range(num_vans):
        child[i] = random.choice([parent1[i], parent2[i]])
    return child

# Mutation Function
def mutate(solution, data, num_vans, mutation_rate=0.2):
    for _ in range(int(len(solution) * mutation_rate)):
        van_from = random.randint(0, num_vans - 1)
        van_to = random.randint(0, num_vans - 1)
        if van_from != van_to and len(solution[van_from]) > 0:
            package_idx = random.choice(solution[van_from])
            solution[van_from].remove(package_idx)
            solution[van_to].append(package_idx)
    return solution

# Genetic Algorithm
def genetic_algorithm(data, num_vans, van_capacity, population_size, generations):
    # Step 1: Initialize Population
    population = initialize_population_knapsack(data, num_vans, van_capacity, population_size)

    for generation in range(generations):
        # Step 2: Evaluate Fitness
        fitness_scores = [calculate_fitness(solution, data) for solution in population]

        # Step 3: Selection
        parents = select_parents(population, fitness_scores)

        # Step 4: Crossover
        new_population = []
        for _ in range(population_size // 2):
            child1 = crossover(parents[0], parents[1], num_vans)
            child2 = crossover(parents[1], parents[0], num_vans)
            new_population.extend([child1, child2])

        # Step 5: Mutation
        mutated_population = [mutate(solution, data, num_vans) for solution in new_population]

        # Replace old population with new one
        population = mutated_population

        # Track Progress
        best_fitness = max(fitness_scores)
        avg_fitness = sum(fitness_scores) / len(fitness_scores)
        print(f"Generation {generation + 1}: Best Fitness = {best_fitness}, Average Fitness = {avg_fitness}")

    # Step 6: Return Best Solution
    final_fitness_scores = [calculate_fitness(solution, data) for solution in population]
    best_solution_index = final_fitness_scores.index(max(final_fitness_scores))
    return population[best_solution_index], max(final_fitness_scores)

# Run the Genetic Algorithm
best_solution, best_fitness = genetic_algorithm(
    filtered_data, NUM_VANS, VAN_CAPACITY, population_size=20, generations=10
)

# Display Final Results
van_stats, total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(
    best_solution, filtered_data
)

print("\nGenetic Algorithm Results with Knapsack Initialization:")
print(f"Total Profit for Deliveries: {total_profit}")
print(f"Total Penalty for Overdue Packages: {total_penalty}")
print(f"Number of Packages Left in Warehouse: {leftover_count}")
print(f"Total Weight of Leftover Packages: {leftover_weight}")
print(f"Total Profit of Leftover Packages: {leftover_profit}")
```
