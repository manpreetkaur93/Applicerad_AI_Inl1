Step 1: Load and Process Data
Tasks:
Load the lagerstatus.csv files.
Validate columns and preprocess the data.

```python
import pandas as pd
import os
import datetime as dt

# File paths
files = [
    "lagerstatus.csv",
    "lagerstatus2.csv",
    "lagerstatus3.csv",
    "lagerstatus4.csv"
]

# Validate file update times
current_time = dt.datetime.now()
for file in files:
    file_mod_time = dt.datetime.fromtimestamp(os.path.getmtime(file))
    if file_mod_time < current_time - dt.timedelta(hours=5):
        raise RuntimeError(f"File {file} is not updated within the last 5 hours.")

# Load and concatenate all files
dataframes = [pd.read_csv(file) for file in files]
filtered_data = pd.concat(dataframes, ignore_index=True)

# Validate required columns
required_columns = ['Paket_id', 'Vikt', 'Förtjänst', 'Deadline']
if not all(col in filtered_data.columns for col in required_columns):
    raise ValueError("One or more required columns are missing!")

# Add Penalty column
filtered_data['Penalty'] = filtered_data['Deadline'].apply(lambda x: -(x**2) if x < 0 else 0)

print("Data loaded and processed successfully!")
```

Step 2: Assign Packages to Vans
Tasks:
Assign packages to 10 vans using a knapsack strategy.
Refine assignments using a genetic algorithm.

### Knapsack Strategy Code:

```python
def assign_packages_with_knapsack(data, num_vans, van_capacity):
    """
    Assign packages to vans using a greedy knapsack approach.
    """
    data['Profit_Weight'] = data['Förtjänst'] / data['Vikt']
    sorted_packages = data.sort_values(by='Profit_Weight', ascending=False).index.tolist()

    solution = [[] for _ in range(num_vans)]
    van_weights = [0] * num_vans

    for package_idx in sorted_packages:
        package_weight = data.loc[package_idx, 'Vikt']
        for van_idx in range(num_vans):
            if van_weights[van_idx] + package_weight <= van_capacity:
                solution[van_idx].append(package_idx)
                van_weights[van_idx] += package_weight
                break

    return solution

# Assign packages
NUM_VANS = 10
VAN_CAPACITY = 800
best_solution = assign_packages_with_knapsack(filtered_data, NUM_VANS, VAN_CAPACITY)
```

### Genetic Algorithm Code:

```python
def calculate_fitness(solution, data):
    """
    Calculates fitness of a solution based on profit and penalty.
    """
    total_profit = 0
    total_penalty = 0

    for van in solution:
        van_profit = sum(data.loc[van, 'Förtjänst']) if isinstance(van, list) else 0
        van_penalty = sum(data.loc[van, 'Penalty']) if isinstance(van, list) else 0
        total_profit += van_profit
        total_penalty += van_penalty

    # Fitness is profit minus penalty
    return total_profit + total_penalty  # Note: penalty is already negative
def genetic_algorithm(data, num_vans, van_capacity, population_size=20, generations=10):
    """
    Refine package assignments using a genetic algorithm.
    """
    population = [assign_packages_with_knapsack(data, num_vans, van_capacity) for _ in range(population_size)]

    for generation in range(generations):
        fitness_scores = [calculate_fitness(solution, data) for solution in population]

        # Select top 2 solutions
        parents = sorted(zip(fitness_scores, population), key=lambda x: x[0], reverse=True)[:2]
        parent1, parent2 = parents[0][1], parents[1][1]

        # Generate new population via crossover and mutation
        new_population = []
        for _ in range(population_size // 2):
            child1 = crossover(parent1, parent2, num_vans)
            child2 = crossover(parent2, parent1, num_vans)
            new_population.extend([mutate(child1, data, num_vans), mutate(child2, data, num_vans)])

        population = new_population

    # Return the best solution
    fitness_scores = [calculate_fitness(solution, data) for solution in population]
    best_solution = population[fitness_scores.index(max(fitness_scores))]
    return best_solution

```

### Step 3: Calculate Results
Tasks:
Compute total profit, penalties, and leftover packages.

```python
def calculate_statistics(solution, data):
    total_profit = 0
    total_penalty = 0
    assigned_packages = []

    for van in solution:
        assigned_packages.extend(van)
        total_profit += sum(data.loc[van, 'Förtjänst'])
        total_penalty += sum(data.loc[van, 'Penalty'])

    leftover_packages = data.index.difference(assigned_packages)
    leftover_weight = sum(data.loc[leftover_packages, 'Vikt'])
    leftover_profit = sum(data.loc[leftover_packages, 'Förtjänst'])

    return total_profit, total_penalty, len(leftover_packages), leftover_weight, leftover_profit

total_profit, total_penalty, leftover_count, leftover_weight, leftover_profit = calculate_statistics(best_solution, filtered_data)
print(f"Total Profit: {total_profit}")
print(f"Total Penalty: {total_penalty}")
print(f"Leftover Packages: {leftover_count}")
print(f"Leftover Weight: {leftover_weight}")
print(f"Leftover Profit: {leftover_profit}")
```

#### Step 4: Visualize Results
Tasks:
Generate histograms for weights, profits, penalties, and leftovers.

```python
import matplotlib.pyplot as plt

van_labels = [f'car_{i+1}' for i in range(NUM_VANS)]
van_weights = [sum(filtered_data.loc[van, 'Vikt']) for van in best_solution]
van_profits = [sum(filtered_data.loc[van, 'Förtjänst']) for van in best_solution]
van_penalties = [sum(filtered_data.loc[van, 'Penalty']) for van in best_solution]

# Van Weight Distribution
plt.figure(figsize=(10, 6))
plt.bar(van_labels, van_weights, color='blue', label='Van Weights')
plt.axhline(VAN_CAPACITY, color='red', linestyle='dashed', label='Van Capacity')
plt.title('Van Weight Distribution')
plt.legend()
plt.show()

# Profit Distribution
plt.figure(figsize=(10, 6))
plt.bar(van_labels, van_profits, color='green', label='Van Profits')
plt.title('Profit Distribution')
plt.legend()
plt.show()

# Penalty Distribution
plt.figure(figsize=(10, 6))
plt.bar(van_labels, van_penalties, color='orange', label='Van Penalties')
plt.title('Penalty Distribution')
plt.legend()
plt.show()

# Leftover Packages
leftover_weights = filtered_data.loc[filtered_data.index.difference([pkg for van in best_solution for pkg in van]), 'Vikt']
plt.figure(figsize=(10, 6))
plt.hist(leftover_weights, bins=20, color='purple', label='Leftover Package Weights')
plt.title('Leftover Package Weight Distribution')
plt.legend()
plt.show()
```

### Step 5: Optimize and Track
Tasks:
Log optimization progress

```python
def track_progress(population, data):
    progress = []
    for solution in population:
        fitness = calculate_fitness(solution, data)
        progress.append(fitness)
    return progress

# Example during optimization
progress = track_progress(population, filtered_data)
plt.plot(progress)
plt.title("Optimization Progress")
plt.xlabel("Generation")
plt.ylabel("Fitness Score")
plt.show()
```
