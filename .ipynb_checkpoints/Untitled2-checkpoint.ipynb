```python
import pandas as pd

def preprocess_data(file_path):
    # Load data
    data = pd.read_csv(file_path)

    # Add penalty column for overdue packages
    data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)

    # Add valid column (valid if Deadline >= 0 and Vikt ≤ 800 kg)
    data['Valid'] = (data['Deadline'] >= 0) & (data['Vikt'] <= 800)

    # Basic statistics
    print("Summary Statistics:")
    print(data.describe())

    return data

# Preprocess each file
file_paths = [
    "lagerstatus.csv",
    "lagerstatus2.csv",
    "lagerstatus3.csv",
    "lagerstatus4.csv"
]

processed_data = [preprocess_data(path) for path in file_paths]
```

```python
import pandas as pd

# Constants
VEHICLE_CAPACITY = 800
NUM_VEHICLES = 10

# Function to allocate packages with refined strategy
def allocate_packages_final(data):
    # Prioritize high penalty, high profit, and light weight
    data['Score'] = (data['Förtjänst'] * 2) - (data['Penalty'] / 5) + (1 / (data['Vikt'] + 1))
    data = data.sort_values(by='Score', ascending=False)

    vehicles = [[] for _ in range(NUM_VEHICLES)]  # Create 10 empty vehicles
    vehicle_weights = [0] * NUM_VEHICLES  # Track current weight of each vehicle
    package_assignment_count = {pkg_id: 0 for pkg_id in data['Paket_id']}  # Track package assignments

    for _, pkg in data.iterrows():
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg['Vikt'] <= VEHICLE_CAPACITY:
                vehicles[i].append(pkg['Paket_id'])
                vehicle_weights[i] += pkg['Vikt']
                package_assignment_count[pkg['Paket_id']] += 1
                break  # Assign package to one vehicle

    # Update Loaded column
    data['Loaded'] = data['Paket_id'].apply(lambda x: package_assignment_count[x])
    return vehicles, data

# Fitness function with heavier penalties for leftovers
def fitness_score(data, vehicles):
    total_profit = 0
    total_penalty = 0
    unallocated_packages = data[data['Loaded'] == 0]

    for vehicle in vehicles:
        for pkg_id in vehicle:
            pkg = data[data['Paket_id'] == pkg_id]
            total_profit += pkg['Förtjänst'].values[0]
            total_penalty += pkg['Penalty'].values[0]

    # Add penalties for unallocated packages
    leftover_weight = unallocated_packages['Vikt'].sum()
    leftover_profit = unallocated_packages['Förtjänst'].sum()
    for _, pkg in unallocated_packages.iterrows():
        total_penalty += pkg['Penalty']

    # Reward for reducing unallocated packages
    reward = len(data) - len(unallocated_packages)

    # Penalize duplicate assignments
    duplicates_penalty = data[data['Loaded'] > 1]['Loaded'].sum() * 10
    total_penalty += duplicates_penalty

    return total_profit + total_penalty + reward * 5, total_penalty, leftover_weight, leftover_profit, unallocated_packages

# Load data and preprocess
data_path = "lagerstatus.csv"  # Path to the first CSV file
data = pd.read_csv(data_path)
data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)

# Allocate packages and calculate fitness score
vehicles, data = allocate_packages_final(data)
fitness, total_penalty, leftover_weight, leftover_profit, unallocated_packages = fitness_score(data, vehicles)

# Save leftover packages to a CSV file
leftover_path = "final_leftover_packages.csv"
unallocated_packages.to_csv(leftover_path, index=False)

# Output results
print("Fitness Score:", fitness)
print("Total Penalty:", total_penalty)
print("Leftover Weight:", leftover_weight)
print("Leftover Profit:", leftover_profit)
for i, vehicle in enumerate(vehicles):
    print(f"Vehicle {i + 1}: {len(vehicle)} packages, total weight: {sum(data[data['Paket_id'].isin(vehicle)]['Vikt'])} kg")

# Save full dataset to output CSV
output_path = "final_output.csv"
data.to_csv(output_path, index=False)
print(f"Results saved to {output_path} with leftover packages!")
#print(f"Leftover packages saved to {leftover_path}")
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Constants
VEHICLE_CAPACITY = 800
NUM_VEHICLES = 10
GENERATIONS = 10  # Ensure loop respects this value
POPULATION_SIZE = 5
MUTATION_COUNT = 5

# Function to allocate packages
def allocate_packages_final(data):
    data['Score'] = (data['Förtjänst'] * 2) - (data['Penalty'] / 5) + (1 / (data['Vikt'] + 1))
    data = data.sort_values(by='Score', ascending=False)

    vehicles = [[] for _ in range(NUM_VEHICLES)]
    vehicle_weights = [0] * NUM_VEHICLES
    package_assignment_count = {pkg_id: 0 for pkg_id in data['Paket_id']}

    for _, pkg in data.iterrows():
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg['Vikt'] <= VEHICLE_CAPACITY:
                vehicles[i].append(pkg['Paket_id'])
                vehicle_weights[i] += pkg['Vikt']
                package_assignment_count[pkg['Paket_id']] += 1
                break

    data['Loaded'] = data['Paket_id'].apply(lambda x: 1 if package_assignment_count[x] > 0 else 0)
    return vehicles, data

# Fitness function
def fitness_score(data, vehicles):
    assigned_packages = set(pkg for vehicle in vehicles for pkg in vehicle)
    data['Loaded'] = data['Paket_id'].apply(lambda x: 1 if x in assigned_packages else 0)

    total_profit = data[data['Loaded'] == 1]['Förtjänst'].sum()
    total_penalty = data[data['Loaded'] == 1]['Penalty'].sum()
    total_penalty += data[data['Loaded'] == 0]['Penalty'].sum()

    reward = len(data) - len(data[data['Loaded'] == 0])
    duplicates_penalty = data[data['Loaded'] > 1]['Loaded'].sum() * 10
    total_penalty += duplicates_penalty

    leftover_weight = data[data['Loaded'] == 0]['Vikt'].sum()
    leftover_profit = data[data['Loaded'] == 0]['Förtjänst'].sum()

    return total_profit + total_penalty + reward * 5, total_penalty, leftover_weight, leftover_profit, data[data['Loaded'] == 0]

# Mutation function
def optimize_mutation(vehicles, data):
    mutated_vehicles = [list(vehicle) for vehicle in vehicles]
    vehicle_weights = [sum(data[data['Paket_id'].isin(vehicle)]['Vikt']) for vehicle in mutated_vehicles]

    leftover_packages = data[data['Loaded'] == 0].sort_values(by='Penalty', ascending=True)
    for _, pkg in leftover_packages.iterrows():
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg['Vikt'] <= VEHICLE_CAPACITY:
                mutated_vehicles[i].append(pkg['Paket_id'])
                vehicle_weights[i] += pkg['Vikt']
                break

    for _ in range(MUTATION_COUNT):
        vehicle_from = random.randint(0, NUM_VEHICLES - 1)
        vehicle_to = random.randint(0, NUM_VEHICLES - 1)

        if mutated_vehicles[vehicle_from]:
            pkg_id = random.choice(mutated_vehicles[vehicle_from])
            pkg_weight = data[data['Paket_id'] == pkg_id]['Vikt'].values[0]
            if vehicle_weights[vehicle_to] + pkg_weight <= VEHICLE_CAPACITY:
                mutated_vehicles[vehicle_from].remove(pkg_id)
                mutated_vehicles[vehicle_to].append(pkg_id)
                vehicle_weights[vehicle_from] -= pkg_weight
                vehicle_weights[vehicle_to] += pkg_weight

    return mutated_vehicles

# Crossover function
def crossover(parent1, parent2, data):
    offspring1 = [list(vehicle) for vehicle in parent1]
    offspring2 = [list(vehicle) for vehicle in parent2]

    crossover_point = random.randint(0, NUM_VEHICLES - 1)
    for i in range(crossover_point, NUM_VEHICLES):
        temp = offspring1[i]
        offspring1[i] = offspring2[i]
        offspring2[i] = temp

    offspring1 = resolve_constraints(offspring1, data)
    offspring2 = resolve_constraints(offspring2, data)

    return offspring1, offspring2

# Constraint resolution
def resolve_constraints(vehicles, data):
    all_assigned_packages = set(pkg for vehicle in vehicles for pkg in vehicle)
    all_packages = set(data['Paket_id'])

    unassigned_packages = list(all_packages - all_assigned_packages)
    package_counts = {pkg: 0 for pkg in all_packages}
    for vehicle in vehicles:
        for pkg in vehicle:
            package_counts[pkg] += 1

    for pkg, count in package_counts.items():
        while count > 1:
            for vehicle in vehicles:
                if pkg in vehicle:
                    vehicle.remove(pkg)
                    count -= 1
                    break

    vehicle_weights = [sum(data[data['Paket_id'].isin(vehicle)]['Vikt']) for vehicle in vehicles]
    for pkg_id in unassigned_packages:
        pkg_weight = data[data['Paket_id'] == pkg_id]['Vikt'].values[0]
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg_weight <= VEHICLE_CAPACITY:
                vehicles[i].append(pkg_id)
                vehicle_weights[i] += pkg_weight
                break

    return vehicles

# Genetic Algorithm with fixed generations
def genetic_algorithm(data):
    population = [allocate_packages_final(data)[0] for _ in range(POPULATION_SIZE)]
    best_solution = None
    best_fitness = -float('inf')

    fitness_progress = []

    for generation in range(GENERATIONS):
        print(f"Generation {generation + 1}")

        fitness_scores = []
        for vehicles in population:
            fitness, total_penalty, _, _, _ = fitness_score(data, vehicles)
            fitness_scores.append((fitness, vehicles))

        fitness_scores.sort(reverse=True, key=lambda x: x[0])

        if fitness_scores[0][0] > best_fitness:
            best_fitness = fitness_scores[0][0]
            best_solution = fitness_scores[0][1]

        fitness_progress.append(best_fitness)

        print(f"Best Fitness in Generation {generation + 1}: {best_fitness}")
        print(f"Top fitness scores: {[int(score[0]) for score in fitness_scores[:3]]}")

        next_generation = [fitness_scores[i][1] for i in range(2)]
        for i in range(0, len(fitness_scores) - 1, 2):
            parent1 = fitness_scores[i][1]
            parent2 = fitness_scores[i + 1][1]
            offspring1, offspring2 = crossover(parent1, parent2, data)
            next_generation.extend([offspring1, offspring2])

        for vehicles in next_generation[:]:
            next_generation.append(optimize_mutation(vehicles, data))

        population = next_generation[:POPULATION_SIZE]

    return best_solution, best_fitness, fitness_progress

# Visualization
def visualize_fitness_progress(fitness_progress):
    generations = list(range(1, len(fitness_progress) + 1))
    plt.figure(figsize=(10, 6))
    plt.plot(generations, fitness_progress, marker='o')
    plt.title("Fitness Progress")
    plt.xlabel("Generation")
    plt.ylabel("Fitness Score")
    plt.grid(True)
    plt.show()

# Load data
data = pd.read_csv("lagerstatus.csv")
data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)

# Run Genetic Algorithm
best_solution, best_fitness, fitness_progress = genetic_algorithm(data)

# Output and visualization
print(f"Final Fitness Score: {best_fitness}")
visualize_fitness_progress(fitness_progress)




#These are moved inside the loop of the genetic_algorithm function, where generation is defined."""
```
