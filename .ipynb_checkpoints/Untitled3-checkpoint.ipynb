```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants
VEHICLE_CAPACITY = 800
NUM_VEHICLES = 10

# Step 1: Read and Preprocess Data
def preprocess_data(file_path):
    # Load CSV data
    data = pd.read_csv(file_path)
    
    # Calculate penalties
    data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)
    
    # Calculate a prioritization score for sorting
    data['Score'] = (data['Förtjänst'] * 2) - (data['Penalty'] / 5) + (1 / (data['Vikt'] + 1))
    data = data.sort_values(by='Score', ascending=False).reset_index(drop=True)
    
    return data

# Visualize initial data distribution
def visualize_initial_data(data):
    plt.figure(figsize=(10, 6))
    plt.hist(data['Vikt'], bins=20, alpha=0.6, label='Weight')
    plt.hist(data['Förtjänst'], bins=20, alpha=0.6, label='Profit')
    plt.title('Initial Data Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Load and preprocess the data
file_path = "lagerstatus.csv"
data = preprocess_data(file_path)

# Visualize the data
visualize_initial_data(data)

# Summary statistics
print("Data Summary:")
print(data.describe())
```

2. **Fitness-utvärdering:**

   - Algoritmen inkluderar mutation och crossover för att skapa nya lösningar varje generation.

7. **Prioritering av försenade paket:**
   - Antalet försenade paket spåras och skrivs ut som del av fitness-utvärderingen, men prioritering kan förstärkas ytterligare.

### Genetisk Algoritm och Fitness-förbättring

#### Stoppkriterium: Algoritmen stannar vid låg förbättring över flera generationer.
####  Policy för försenade paket: Försenade paket prioriteras dynamiskt.
####  Skalbarhet: Kan bearbeta stora filer genom chunksize.







```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Konstanter
VEHICLE_CAPACITY = 800
NUM_VEHICLES = 10
GENERATIONS = 10
POPULATION_SIZE = 5
MUTATION_COUNT = 5

# Funktion: Ladda stora dataset
def load_large_data(file_path, chunksize=10000):
    data_chunks = pd.read_csv(file_path, chunksize=chunksize)
    data = pd.concat(data_chunks, ignore_index=True)
    data['Penalty'] = data['Deadline'].apply(lambda x: -x**2 if x < 0 else 0)
    return data

# Funktion: Tilldela paket med strikt policy för försenade paket
def allocate_packages_with_deadline_policy(data):
    data['Score'] = (
        (data['Förtjänst'] * 2) - 
        (data['Penalty'] / 3) + 
        (1 / (data['Vikt'] + 1)) + 
        (data['Deadline'] < 0) * 5  # Försenade paket får högre vikt
    )
    data = data.sort_values(by='Score', ascending=False)

    vehicles = [[] for _ in range(NUM_VEHICLES)]
    vehicle_weights = [0] * NUM_VEHICLES
    package_assignment_count = {pkg_id: 0 for pkg_id in data['Paket_id']}

    for _, pkg in data.iterrows():
        for i in range(NUM_VEHICLES):
            if vehicle_weights[i] + pkg['Vikt'] <= VEHICLE_CAPACITY:
                vehicles[i].append(pkg['Paket_id'])
                vehicle_weights[i] += pkg['Vikt']
                package_assignment_count[pkg['Paket_id']] += 1
                break

    data['Loaded'] = data['Paket_id'].apply(lambda x: 1 if package_assignment_count[x] > 0 else 0)
    return vehicles, data

# Funktion: Fitness-utvärdering
def fitness_score(data, vehicles):
    assigned_packages = set(pkg for vehicle in vehicles for pkg in vehicle)
    unallocated_packages = data[~data['Paket_id'].isin(assigned_packages)]
    delayed_delivered = len(data[(data['Paket_id'].isin(assigned_packages)) & (data['Deadline'] < 0)])
    delayed_stock = len(data[~data['Paket_id'].isin(assigned_packages) & (data['Deadline'] < 0)])
    total_profit = data[data['Paket_id'].isin(assigned_packages)]['Förtjänst'].sum()
    total_penalty = unallocated_packages['Penalty'].sum()

    return total_profit - total_penalty, total_penalty, delayed_delivered, delayed_stock

# Funktion: Mutation
def optimize_mutation(vehicles, data):
    mutated_vehicles = [list(vehicle) for vehicle in vehicles]
    vehicle_weights = [
        sum(data[data['Paket_id'].isin(vehicle)]['Vikt']) for vehicle in mutated_vehicles
    ]

    for _ in range(MUTATION_COUNT):
        vehicle_from = random.randint(0, NUM_VEHICLES - 1)
        vehicle_to = random.randint(0, NUM_VEHICLES - 1)

        if mutated_vehicles[vehicle_from]:
            pkg_id = random.choice(mutated_vehicles[vehicle_from])
            pkg_weight = data[data['Paket_id'] == pkg_id]['Vikt'].values[0]

            if vehicle_weights[vehicle_to] + pkg_weight <= VEHICLE_CAPACITY:
                mutated_vehicles[vehicle_from].remove(pkg_id)
                mutated_vehicles[vehicle_to].append(pkg_id)
                vehicle_weights[vehicle_from] -= pkg_weight
                vehicle_weights[vehicle_to] += pkg_weight

    return mutated_vehicles

# Funktion: Genetisk Algoritm med Stoppkriterium
def genetic_algorithm_with_stopping(data, max_generations=GENERATIONS, tolerance=0.01, patience=5):
    population = [allocate_packages_with_deadline_policy(data)[0] for _ in range(POPULATION_SIZE)]
    best_solution = None
    best_fitness = -float('inf')
    fitness_progress = []
    stagnant_generations = 0

    for generation in range(max_generations):
        fitness_scores = []
        for vehicles in population:
            fitness, total_penalty, delayed_delivered, delayed_stock = fitness_score(data, vehicles)
            fitness_scores.append((fitness, vehicles, total_penalty, delayed_delivered, delayed_stock))

        fitness_scores.sort(reverse=True, key=lambda x: x[0])

        if fitness_scores[0][0] > best_fitness:
            best_fitness = fitness_scores[0][0]
            best_solution = fitness_scores[0][1]
            best_total_penalty = fitness_scores[0][2]
            best_delayed_delivered = fitness_scores[0][3]
            best_delayed_stock = fitness_scores[0][4]
            stagnant_generations = 0
        else:
            stagnant_generations += 1

        fitness_progress.append(best_fitness)

        if stagnant_generations >= patience:
            print(f"Algoritmen stannade vid generation {generation + 1} på grund av låg förbättring.")
            break

        next_generation = [fitness_scores[i][1] for i in range(POPULATION_SIZE // 2)]
        for i in range(0, len(next_generation), 2):
            if i + 1 < len(next_generation):
                parent1, parent2 = next_generation[i], next_generation[i + 1]
                offspring1, offspring2 = optimize_mutation(parent1, data), optimize_mutation(parent2, data)
                next_generation.extend([offspring1, offspring2])

        population = next_generation[:POPULATION_SIZE]

    return best_solution, best_fitness, fitness_progress, best_total_penalty, best_delayed_delivered, best_delayed_stock

# Visualisering av viktfördelning
def visualize_vehicle_allocation(vehicles, data, best_fitness, total_penalty, delayed_delivered, delayed_stock):
    vehicle_weights = [
        sum(data[data['Paket_id'].isin(vehicle)]['Vikt']) for vehicle in vehicles
    ]

    plt.figure(figsize=(12, 8))
    plt.bar(range(1, NUM_VEHICLES + 1), vehicle_weights, color='skyblue', alpha=0.7)
    plt.axhline(y=VEHICLE_CAPACITY, color='r', linestyle='--', label="Max Kapacitet")
    plt.title("Viktfördelning per fordon")
    plt.xlabel("Fordon")
    plt.ylabel("Totalvikt (kg)")
    plt.legend()
    plt.show()

    print(f"\nFitness Score för bästa lösningen: {best_fitness}")
    print(f"Total straffavgift: {total_penalty}")
    print(f"Antal försenade paket (levererade): {delayed_delivered}")
    print(f"Antal försenade paket (i lager): {delayed_stock}")

# Visualisering av fitnessprogress
def visualize_fitness_progress(fitness_progress):
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(fitness_progress) + 1), fitness_progress, marker='o', label='Fitness Score')
    plt.title('Fitness Progress Over Generations')
    plt.xlabel('Generation')
    plt.ylabel('Fitness Score')
    plt.grid(True)
    plt.legend()
    plt.show()

# Ladda data
data = load_large_data("lagerstatus.csv")

# Kör algoritmen
best_solution, best_fitness, fitness_progress, total_penalty, delayed_delivered, delayed_stock = genetic_algorithm_with_stopping(data)

# Visualiseringar
visualize_fitness_progress(fitness_progress)
visualize_vehicle_allocation(best_solution, data, best_fitness, total_penalty, delayed_delivered, delayed_stock)
```
